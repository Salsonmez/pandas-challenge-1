## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
df[['qty', 'unit_price', 'line_subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['shipping_price'] = df['unit_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)
df[['order_id', 'unit_weight', 'shipping_price']].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

df['sales_tax'] = 0.0925
df['total_price'] = df['line_subtotal'] + df['shipping_price']
df['total_price_with_sales_tax'] = df['total_price'] * (1 + df['sales_tax'])

df[['qty', 'unit_price', 'line_subtotal', 'unit_weight', 'shipping_price', 'total_price_with_sales_tax']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

df['cost_each_line'] = df['unit_price'] * df['qty'] + df['shipping_price']
df.head()

```

```python
# Create a column for the profit of each line using line cost and line price

df['profit'] = df['total_price_with_sales_tax'] - df['cost_each_line']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
total_spending = df.groupby('client_id')['cost_each_line'].sum()
top_clients_by_spending = total_spending.nlargest(5)
top_clients_by_spending
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

summary_dfs = []
top_clients_ids = [78965, 85931, 38473, 86721, 84344]
for client_id in top_clients_ids:

    client_data = df[df['client_id'] == client_id]
    total_units_purchased = client_data['qty'].sum()
    total_shipping_price = client_data['shipping_price'].sum()
    total_revenue = client_data['total_price_with_sales_tax'].sum()
    total_profit = client_data['profit'].sum()
    
    summary_dfs.append({
        'Client': client_id,
        'Total Units Purchased': total_units_purchased,
        'Total Shipping Price': total_shipping_price,
        'Total Revenue': total_revenue,
        'Total Profit': total_profit
    })

summary_df = pd.DataFrame(summary_dfs)

summary_df
```
